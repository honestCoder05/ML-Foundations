#### Intro to Linear Algebra

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 40, 1000) # create the time space line
```

##### Distance travelled by the robber and sheriff, respectively

```python
d_r = 2.5 * t # since robber is travelling at 150km/hour
d_s = 3 * (t - 5) # since sheriff is travelling at 180km/hour
```

```python
fig, ax= plt.subplots()
plt.title("Bank robber caught time and distance")
plt.xlabel("time (in minutes)")
plt.ylabel("distance (in km)")
ax.set_xlim([0, 40])
ax.set_ylim([0, 100])
ax.plot(t, d_r, c = "green")
ax.plot(t, d_s, c = "brown")
plt.axvline(x = 30, color = "purple", linestyle = "--")
_ = plt.axhline(y = 75, color = "purple", linestyle = "--")
```
